```python
#count of each transport mode
cyclist = [c for c in counts.columns if "Cyclist" in c]

pedestrian = [c for c in counts.columns if "Pedestrian" in c]

scooter = [c for c in counts.columns if "Scooter" in c]

print(len(cyclist))

print(len(pedestrian))

print(len(scooter))
```

```python
#missing values
missing = counts.isnull().sum()

missing = missing[missing > 0]

missing.sort_values(ascending=False)
```

```python
#check possible date range
#exact 4 years of data - jan 2022 till dec 2025
counts["Date"] = pd.to_datetime(counts["Date"])

print(counts["Date"].min())

print(counts["Date"].max())
```
